```python
import spacy
```

```python
from spacy import displacy
```

```python
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
dir = "booksummaries.csv"
```

```python
df = pd.read_csv(dir)
```

```python
samenvattingen = df["plot"][:5]
```

```python
for samenvatting in samenvattingen:
    doc = nlp(samenvatting)
    displacy.render(doc, style="ent", jupyter=True)
```
